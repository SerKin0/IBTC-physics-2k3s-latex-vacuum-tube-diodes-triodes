```python
import matplotlib.pyplot as plt
import numpy as np

def smart_millimeter_plot(x, y, fig_scale=1.0):
    """
    Автоматически подстраивает миллиметровку под данные
    """
    # Определяем диапазон данных
    x_range = np.max(x) - np.min(x)
    y_range = np.max(y) - np.min(y)
    
    # Автоматический выбор шага сетки
    major_step = 10 ** np.floor(np.log10(max(x_range, y_range))) / 2
    minor_step = major_step / 5
    
    # Автоматический размер графика
    fig_width = max(8, x_range / major_step * 1.5) * fig_scale
    fig_height = max(6, y_range / major_step * 1.5) * fig_scale
    
    fig, ax = plt.subplots(figsize=(fig_width, fig_height))
    
    # Настройка сетки
    ax.grid(True, which='major', linestyle='-', linewidth=0.7, alpha=0.4)
    ax.grid(True, which='minor', linestyle='--', linewidth=0.3, alpha=0.2)
    
    ax.xaxis.set_major_locator(plt.MultipleLocator(major_step))
    ax.yaxis.set_major_locator(plt.MultipleLocator(major_step))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(minor_step))
    ax.yaxis.set_minor_locator(plt.MultipleLocator(minor_step))
    
    return fig, ax

# Пример
x = np.random.normal(0, 10, 100)
y = 2*x + 3 + np.random.normal(0, 5, 100)

fig, ax = smart_millimeter_plot(x, y, fig_scale=0.8)
ax.scatter(x, y, alpha=0.7)
ax.set_xlabel('X значения')
ax.set_ylabel('Y значения')
ax.set_title('Точечный график на адаптивной миллиметровке')
plt.show()
```
